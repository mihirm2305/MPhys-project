```python
# import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load Dataset1.csv
# df = pd.read_csv('Dataset1.csv') - no headers
df = pd.read_csv('Dataset1.csv', header=None)
df.head()
```

```python
x = df[0]
y = df[1]
```

```python
# plot the data
plt.scatter(x, y)
```

```python
# fit the region x < 3 to A*x^n (with initial guess n=0.5)

from scipy.optimize import curve_fit

def func(x, A, n):
    return A*x**n

popt, pcov = curve_fit(func, x[x<3.6], y[x<3.6], p0=[1, 0.5])

# plot the fit
xfit = np.linspace(0, 3.6, 100)
yfit = func(xfit, *popt)
plt.plot(xfit, yfit, 'r-')
plt.scatter(x, y)
plt.xlabel('T(K)')
plt.ylabel('Ce/T (mJ/mol K^2)')

# annotate the plot with the fit equation
plt.text(2, 0.5, f'fit equation: $y = {popt[0]:.2f} \\cdot x^{{{popt[1]:.2f}}}$', fontsize=12, color='red')

# print the fit parameters
print('A =', popt[0])
print('n =', popt[1])
```
